```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the Dataset
dataset = pd.read_csv('Chronic kidney disease prediction.csv')
```

```python
dataset
```

```python
#it replaces 'yes' with 1 and 'no' with 0.
dataset['Possible'] = dataset['Possible'].replace(['yes','no'], [1,0])
```

```python
dictonary = {
        "rbc": {
        "abnormal":1,
        "normal": 0,
    },
        "pc":{
        "abnormal":1,
        "normal": 0,
    },
        "pcc":{
        "present":1,
        "notpresent":0,
    },
        "ba":{
        "notpresent":0,
        "present": 1,
    },
        "htn":{
        "yes":1,
        "no": 0,
    },
        "dm":{
        "yes":1,
        "no":0,
    },
        "cad":{
        "yes":1,
        "no": 0,
    },
        "appet":{
        "yes":1,
        "poor": 0,
    },
        "pe":{
        "yes":1,
        "poor":0,
    },
        "ane":{
        "yes":1,
        "no":0,
    },
        "sg":{
        "a":0,
        "b":1,
        "c":2,
        "d":3,
        "e":4,
        }
}
```

```python
dataset=dataset.replace(dictonary)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["Possible"].value_counts()
```

```python
indep=dataset[['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu',
       'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane',]]
dep=dataset['Possible']
```

```python
indep.shape
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier

param_grid = { 
    'criterion':['friedman_mse', 'squared_error'],
    'n_estimators': [50, 100],  
    'max_features': ['sqrt', 'log2', None]
}


# Create GridSearchCV with the classifier instance and parameter grid
grid = GridSearchCV(GradientBoostingClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1')

# Fit the grid search
grid.fit(X_train, Y_train)
```

```python
# print best parameter after tuning 
#print(grid.best_params_) 
re=grid.cv_results_
#print(re)
grid_predictions = grid.predict(X_test) 
   

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, grid_predictions)



# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, grid_predictions)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
